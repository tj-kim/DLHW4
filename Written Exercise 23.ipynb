Perform the same exercise as above on adversarial examples. Specifically, are adversarial examples more
similar to natural examples of the same class in the input space or in the feature space?



```python
import hw4_part3
import numpy as np

import hw4_part1
import hw4_part2

import hw4_utils
import hw4_mnist

# Load Data
mnist_data = hw4_mnist.load_data()
train_X = mnist_data.train.X
train_Y = mnist_data.train.Y
```

### Loop through values and find 100 similarities

Make sure to attack for each integer when attacking.

```python
m = hw4_part3.MNISTModelRegular()
m.load()
```

```python
integer = np.arange(10)
sample_size = 10
n = 10 # number of maximums to take

input_space_sim = []
feature_space_sim = []

flattened_X = np.squeeze(
    np.reshape(train_X,[train_X.shape[0],train_X.shape[1]*train_X.shape[2],train_X.shape[3]])
    ,axis = 2)

# Loop through different integers
for i in integer:
    idx = np.where(train_Y[0:300]==i)[0]
    
    # Attack 
    # CWL2 on M1
    att = hw4_part1.CWL2Attacker(
            model=m,
            num_steps=10,
            target=None,
            learning_rate=1.0,
            learning_rate_decay=0.9,
            k=0,
            c=4
        )
    
    att_X = att.attack(train_X[0:320],train_Y[0:320])
    att_X_flattened = np.squeeze(
        np.reshape(att_X,[att_X.shape[0],att_X.shape[1]*att_X.shape[2],att_X.shape[3]])
        ,axis = 2)

    
    # Loop through different instance of same integer
    for j in np.arange(sample_size):
        print("doing number",i,j)
        interest_X = np.expand_dims(att_X_flattened[idx[j]],axis=0)
        
        # Find similarity at input space, take 10 maximum values
        temp_input_sim = np.squeeze(np.moveaxis(np.matmul(interest_X, flattened_X.T), 0, 1),axis=1)
        max_ind = (-temp_input_sim).argsort()[:n]
        
        count = 0
        
        for k in max_ind:
            if i == train_Y[k]:
                count += 1/n
        
        input_space_sim += [count]
        
        
        # Find similarity at feature space, take 10 maximum values
        rep = hw4_part3.Representer(m, np.expand_dims(att_X[idx[j]],axis=0)
                                    ,np.expand_dims(train_Y[idx[j]],axis=0))
        temp_input_sim = rep.similarity(train_X)
        
        max_ind = (-temp_input_sim).argsort()[:n]
        
        count = 0
        
        for k in max_ind:
            if i == train_Y[k]:
                count += 1/n
        
        feature_space_sim += [count]
```

```python
print("input",np.mean(input_space_sim))
print("feature",np.mean(feature_space_sim))
```
